 # Bar graph

```python


from matplotlib import pyplot as plt
import numpy as np
company=['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit=[40, 2, 34, 12]

xpos=np.arange(len(company))
plt.xticks(xpos,company,rotation='vertical')#?
plt.ylabel('revenue')
plt.title('US Stocks')
plt.bar(xpos-0.1,revenue,label='Revenue',width=0.3)
plt.bar(xpos+0.2,profit,label='Profit',width=0.2
       )
plt.margins(0.1)
plt.legend()
plt.style.use('_mpl-gallery')
plt.show()
```

```python
auto=pd.read_csv("Automobile_data.csv")
gauto=auto.groupby('company')
y=gauto.price.max()
x=gauto['company'].max()
```

```python
plt.plot(x,y,label='Cars',linewidth='3')
plt.title('CARS')
plt.xlabel('COMPANY NAME')
plt.ylabel('MAX PRICE')
plt.legend()

plt.xticks(x,rotation=90)
plt.rcParams['figure.figsize']=(30,10)
plt.tight_layout()
plt.show()
```

```python
## if given more than one value to a tick then it will take the max value of that value of that tick
x=auto.company
y=auto['length']
# gauto['length'].max()
plt.plot(x,y)
plt.xticks(x,rotation=90)
plt.show()
```

```python
x=gauto.company.max()
y2=gauto.price.max()
y1=gauto.horsepower.max()
plt.plot(x,y2,label='Price max',linewidth='1')
plt.plot(x,y1,label='Horsepower',linewidth='3')

plt.title('CARS')
plt.xlabel('COMPANY NAME')
plt.ylabel('Data')
plt.legend()
plt.style.use('Solarize_Light2')
plt.xticks(x,rotation=90)

plt.rcParams["figure.figsize"] = (30, 3)
plt.rcParams['font.size']=30

plt.show()
```

```python
company=gauto.company.min()
min_mileage=gauto['average-mileage'].min()
max_mileage=gauto['average-mileage'].max()
avg_mileage=gauto['average-mileage'].mean()
price=gauto.price.min()/1000
xpos=np.arange(len(company))
plt.xticks(xpos,company,rotation=90)
plt.bar(xpos-0.2,min_mileage,label="Min_mileage",width=0.1)
plt.bar(xpos,min_mileage,color='black',label="Min_mileage",width=0.1,alpha=0.8)
plt.bar(xpos+0.2,avg_mileage,color='red',label="Avg_mileage",width=0.2)
plt.bar(xpos,price,color='yellow',label="Avg_mileage",width=0.7,alpha=0.4)
plt.legend()
plt.title('CARS')
plt.xlabel('COMPANY NAME')
plt.ylabel('Mileage')
plt.rcParams['figure.figsize']=(30,10)# to increase fig size
plt.rcParams['font.size']=30# to increase font size
plt.style.use('Solarize_Light2')
plt.show()
```
